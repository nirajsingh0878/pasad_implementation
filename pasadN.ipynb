```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('swat_train.csv')
# header=None,False
df['Normal/Attack'].loc[(df['Normal/Attack'] == 'Normal')] = 0
df['Normal/Attack'].loc[(df['Normal/Attack'] == 'Attack')] = 1
print(df.shape)
df.head()
```

```python
#remove timestamp 
data = df.iloc[:,1:]
print(data.shape)
```

```python
# Normal dataset
data_n = data[data['Normal/Attack']==0]
print(data_n.shape)
data_n.head()
```

```python
# testing dataset
data_a = data[data['Normal/Attack']==1]
print(data_a.shape)
data_a.head()
```

```python
#create only 5000 row normal dataset of swat for training
import math
swatN = data_n.iloc[:5000,:]
# swatN = data_n.iloc[:,:]
N,M = swatN.shape
L = math.floor(N/2)
K=N-L
print(swatN.shape)
print(swatN['FIT101'])
```

```python
#create only 5000 row attack dataset of swat for testing
import math
swatA = data_a.iloc[:5000,:]
# swatA = data_a.iloc[:,:]
N,M = swatA.shape
L = math.floor(N/2)
K=N-L
print(swatA.shape)
print(swatA['FIT101'])
```

```python
from scipy.linalg import hankel
def pasad_train(T,N,L):
    print("N:"+str(N)+ " L:"+str(L)+ " K:"+str(K))
     
    ROW  = T[:L]
    COLUMN = T[L:]
    print("Row vector:"+str(ROW.shape))
    print("Column vector:"+str(COLUMN.shape))
    
    r_mat = hankel(ROW,COLUMN)
    u, s, vh = np.linalg.svd(r_mat)  
    #r < L 
    r = math.floor(L/2)
    ut = np.transpose(u[:,:r])
    return ut
```

```python
###################################### ALGORITHM ############################################3
#step 1 creating hankel 
#step 2 eigenvector uT matrix creating 

ut = pasad_train(swatN.iloc[:,0:1],N,L)
print(ut.shape)
print(type(ut))
print(ut)    
```

```python
# step 3 and 4
temp = swatN.iloc[:,0:1]
x = hankel(temp[:L],temp[L:])
print(x.shape)
s = x[:,0]

for i in range(1,K):
    s=s+x[:,i]
# print(s)

c = s/K
# print(c)

#second learning parameter
ctilde = ut.dot(c)
print(ctilde)
print(ctilde.shape)
```

```python
# NOW CALCULATE DMAX FROM TRAINING SAMPLE FOR SETTING THETA FOR THRESHOLD
d=np.zeros(K)
t= swatN.iloc[:,0:1]
T = hankel(t[:L],t[L:])
print(T.shape)
for j in range(0,K-1):
    d[j]=np.linalg.norm(ctilde-ut.dot(T[:,j]))
dmax=np.amax(d)
print(dmax.shape)
print(dmax)
```

```python
#NOW RUN SAMPLE FOR ATTACK DATASET FOR TESTING 
temp = swatA.iloc[:,0:1]
xj = hankel(temp[:L],temp[L:])
print(xj.shape) 
print(xj)
for i in range(1,K):
    y = ctilde - ut.dot(xj[:,i])
    dj = y.dot(np.transpose(np.transpose(y)))
    if(dj>dmax):
        print(str(dmax)+"  "+str(dj))
        print("Alarm!Value Exceeded")
```
